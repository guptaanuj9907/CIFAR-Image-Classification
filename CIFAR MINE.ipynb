# DATA PREPARATION

```python
x_train=x_train.astype("float32")
x_test=x_test.astype("float32")
```

```python
x_train
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
x_train
```

```python
y_train=keras.utils.to_categorical(y_train, num_classes=10)
y_test=keras.utils.to_categorical(y_test, num_classes=10)
```

```python
y_train
```

```python
x_train.shape
```

```python
Input_shape=x_train.shape[1:]
Input_shape
```

# TRAIN MODEL

```python
from keras.models import Sequential
from keras.layers import Conv2D,MaxPool2D,AveragePooling2D,Dropout,Dense,Flatten
from keras.optimizers import Adam
from keras.callbacks import tensorboard_v1
```

```python
model=Sequential()
model.add(Conv2D(32,kernel_size=(3,3),activation='relu',input_shape=Input_shape))
model.add(Conv2D(32,kernel_size=(3,3),activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(0.3))

model.add(Conv2D(64,kernel_size=(3,3),activation='relu')) #64 is no of feature detector
model.add(Conv2D(64,kernel_size=(3,3),activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(0.2))

model.add(Flatten())

model.add(Dense(units=512,activation='relu'))
model.add(Dense(units=512,activation='relu'))
model.add(Dense(units=10,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(x_train,y_train,epochs=1,batch_size = 32,shuffle=True)
```

```python
x_train.shape,y_train.shape
```

```python
evaluation=model.evaluate(x_test,y_test)
```

```python
print(evaluation)
```

```python
print("Test Accuracy:{}".format(evaluation[1]))
```

```python
prediction=model.predict_classes(x_test)
```

```python
prediction
```

```python
y_test=y_test.argmax(1)
```

```python
width=5
height=5
fig,axes=plt.subplots(width,height,figsize=(25,25))
axes=axes.ravel() #ravel is ued to make contious 1-D array
for i in range(width*height):
    index=np.random.randint(len(x_test))
    axes[i].imshow(x_test[index])
    axes[i].set_title("predicted:{}\n Actual:{}".format(prediction[index],y_test[index]))
    axes[i].axis('off')
plt.subplots_adjust(hspace=0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,prediction)
cm
```

```python
plt.figure(figsize = (10, 10))
sns.heatmap(cm,annot = True)
```
